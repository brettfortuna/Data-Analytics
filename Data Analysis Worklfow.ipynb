```python
con = sqlite3.connect(r'C:\Users\homer\AppData\Roaming\DBeaverData\workspace6\.metadata\sample-database-sqlite-1\Chinook.db')

music1 = pd.read_sql_query("select a.name as artist \
,aa.title as album_title \
,t.name as track \
,t.composer \
,g.name as genre \
,il.InvoiceId \
,t.unitprice as price\
,il.unitprice as price_sold \
,il.quantity \
from Artist a \
left join Album aa on a.artistid = aa.ArtistId \
left join Track t on t.albumid = aa.albumid \
left join Genre g on g.genreid = t.GenreId \
left join Invoiceline il on il.trackId = t.trackid",con)
con.close()
```

```python
#sum the sales and quantity sold by artists, 
music2= music1[['artist', 'album_title', 'Composer', 'genre', 'price_sold', 'Quantity']]
music2
```

```python
#using Groupby
music3 = pd.DataFrame(music2.groupby(['artist', 'album_title', 'Composer', 'genre'])['price_sold'].sum()).reset_index()
#music3 = pd.DataFrame(music2.groupby(['artist', 'album_title', 'Composer', 'genre'])['price_sold','Quantity'].sum()).reset_index()
music3
```

```python
music4 = pd.DataFrame(music2.groupby(['artist', 'album_title','Composer','genre'])['price_sold','Quantity'].sum()).reset_index()
music4.head()
```

```python
music5 = pd.DataFrame(music2.groupby(['artist', 'album_title','genre']).aggregate({'price_sold':'sum', 'Quantity':'count'})).reset_index()
music5
```

```python
#finding the sales with a total_price >$10
music_10 = music5[(music5.price_sold > 10)]
music_10
```

```python
#Find the sales for Rock and Total quanity sold >10
genre_q = music5[(music5.Quantity > 10) & (music5.genre =='Rock')]
genre_q
```

```python
#Find total artist with no sales
nosales = music5[(music5.Quantity == 0)]
nosales
```
